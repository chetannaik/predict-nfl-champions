#DATA MUNGING

We are collecting the data from 2002 to 2013 from Sunshine forecast (http://www.repole.com/sun4cast/data.html) and the recent season (2014) data from (http://www.drwagpicks.com). The functions that are used to extract these data and merge is described below. 

+ **get_all_year_data()**

    * Getting yearwise previous data from the CSV files and store them into a temporary DataFrame called **df**.
    * df contains 5 coloumns named *'date','away_team','home_team','away_score', 'home_score', that is being extracted from the csv file data which are named 'Date', 'Home Team', 'Visitor', 'Home Score', 'Visitor Score' in the csv. 
    * Adding a synthetic coloumn named 'winner' by comparing the 'Home Score' and 'Visitor Score'. 
    * Appending them into a list called **fileList**. 
    * Repeating these steps for all the year from 2002 to 2013 and adding them into a final data frame named **gamesDF**.
    * Storing the previous years game information in *prevGamesDF*.
    
+ **get_recent_season_data()**

    * Doing the same steps as in the function get_all_year_data but for 2014 data.
    * Storing it into *recentGamesDF*.

+ Merging *prevGamesDF* and *recentGamesDF* to get *gamesDF* containg game information from 2002 to 2014.

+ **get_data(gamesDF)**

    * Converting the *gamesDF* DataFrame in the format required (dictionary) by the baseline model. 
   
+ Dumping all the data into a pickle file to reuse it in a faster way rather than creating DataFrames again and again for every run.

```python
# A utility function to compare the results side-by-side
def side_by_side(*objs, **kwds):
    from pandas.core.common import adjoin
    space = kwds.get('space', 4)
    reprs = [repr(obj).split('\n') for obj in objs]
    print adjoin(space, *reprs)
```

```python
def get_all_year_data():
    """ Reads the CSV files and convert them into pandas dataFrame in a specified format.
   
    Args: 
        CSV files of previous year (2002 to 2013)
    
    Returns:
        gamesDF dataframe consisting of all the data from 2002 to 2013"""
    
    # Reading the csv files from the specified folder
    path = r'/media/ankit/Stuffs/Courses/CSE591 - Data Science/College Football Data/CFB_lines'
    allFiles = glob.glob(path + "/*.csv")
    
    fileList = []
    for f in allFiles:
        df = pd.read_csv(f)
            # Getting Selected Coloumns
        df = df[['Date', 'Home Team', 'Visitor', 'Home Score', 'Visitor Score']]

        # Changing the header names
        df.rename(columns={'Date':'date',
                           'Home Team':'home_team',
                           'Visitor': 'away_team',
                           'Home Score': 'home_score',
                           'Visitor Score': 'away_score'}, inplace=True)

        # Replacing + symbol in names with ' '

        df['home_team'] = df['home_team'].map(lambda x: x.replace('+','_'))
        df['away_team'] = df['away_team'].map(lambda x: x.replace('+', '_'))



        # Creating Years Column
        def get_year(col):
            return col.split('/')[-1]

        df['year'] = df['date'].apply(get_year)
        t =  min(df.year)
        df['year'] = int(t)
        
        # changing data type pf home_score and away_score
        def change_type(col):
            try:
                col = float(col)
            except:
                col = 0
            return col
            
        df['home_score'] = df['home_score'].apply(lambda x: change_type(x))
        df['away_score'] = df['away_score'].apply(lambda x: change_type(x))
        
        # Creating winner column 
        def get_winner(home_score, away_score, home_team, away_team):
            if home_score > away_score:
                return home_team
            else:
                return away_team

        df['winner'] = df.apply(lambda x: get_winner(x['home_score'], x['away_score'], x['home_team'], x['away_team']), axis=1)
        fileList.append(df)
    gamesDF = pd.concat(fileList)
    gamesDF = gamesDF[['year','date','away_team','home_team','away_score', 'home_score', 'winner']]
    
    
    return gamesDF
```

```python
def get_recent_season_data():
    """ Gets the gamesDF dataframe which contains the previous year play data and adds the 2014 data.
    
    Args:
        CSV files with 2014 games information
    
    Returns: 
        gamesDF with same coloumns but having the 2014 data.
    """
    path = r'/media/ankit/Stuffs/Courses/CSE591 - Data Science/College Football Data/cfblines2014'
    allFiles = glob.glob(path + "/*.csv")
    
    fileList = []
    for f in allFiles:
        df = pd.read_csv(f)
            # Getting Selected Coloumns
        df = df[['Date','Home Team', ' Vis Team', ' Score', ' Vscore']]

        # Changing the header names
        df.rename(columns={'Date':'date',
                           'Home Team':'home_team',
                           ' Vis Team': 'away_team',
                           ' Score': 'home_score',
                           ' Vscore': 'away_score'}, inplace=True)

        # Replacing + symbol in names with ' '

        df['home_team'] = df['home_team'].map(lambda x: x.replace('+','_').lstrip())
        df['away_team'] = df['away_team'].map(lambda x: x.replace('+', '_').lstrip())



        # Creating Years Column
        df['year'] = 2014
        
        # changing data type pf home_score and away_score
        def change_type(col):
            try:
                col = float(col)
            except:
                col = 0
            return col
            
        df['home_score'] = df['home_score'].apply(lambda x: change_type(x))
        df['away_score'] = df['away_score'].apply(lambda x: change_type(x))
        
        # Creating winner column 
        def get_winner(home_score, away_score, home_team, away_team):
            if home_score > away_score:
                return home_team
            else:
                return away_team

        df['winner'] = df.apply(lambda x: get_winner(x['home_score'], x['away_score'], x['home_team'], x['away_team']), axis=1)
        fileList.append(df)
    gamesDF = pd.concat(fileList)
    gamesDF = gamesDF[['year','date','away_team','home_team','away_score', 'home_score', 'winner']]
    
    
    return gamesDF
```

```python
def get_data(gamesDF):
    """Gets the data in the format required by the baseline model.
  
    Args:
        allYearData: A pandas dataframe containing games data with following columns.
            'year',
            'date',
            'away_team',
            'home_team',
            'away_score',
            'home_score',
            'winner'
        
    Returns:
        A dictionary which containes data in the format that baseline model expects.
    """
    years = set(gamesDF.year)
    collegeBaselineData = dict()
    for year in years:
        homeTeams = set(list(set(gamesDF[gamesDF.year == year].home_team)) + list(set(gamesDF[gamesDF.year == year].away_team)))
        homeData = dict()
        for homeTeam in homeTeams:
            awayData = dict()
            for ix, row in enumerate(gamesDF.iterrows()):
                if row[1].year == year and row[1].home_team == homeTeam:
                    away_team = row[1].away_team
                    home_score = row[1].home_score
                    away_score = row[1].away_score
                    winner = row[1].winner
                    awayData[away_team] = [home_score, away_score, winner]
            homeData[homeTeam] = awayData
        collegeBaselineData[year] = homeData
    return collegeBaselineData
```

#BASELINE MODEL

This baseline model is a *"Point Score Difference Model"*. In this model we take the point difference (difference of total point scored and total points allowed) to calculate the probability of a team to win. In mathematical representation, point score difference model can be written as:

\begin{equation}
M.r = p
\end{equation}


where,<br />
$M$ is  a matrix of size (#Teams)x(#Teams)<br />
$r$ is a vector of size #Teams.<br />
$p$ is a vector of size #Teams. 


**Note** - The number of teams have been for college football is different for different years. So, the size of the matrix and vectors used in this model is not fixed.

- $M$ has the number of games played by each team on the diagonal, and the negation of the number of times each team played head-to-head on the off-diagonal spots.

- $r$ vector is a list of the average end-of-regular season point differentials for each team.

- $p$ is our prediction vector. 

```python
def get_avg_team_score(data):
    """Calculates average team score for a given year.

    Given yearly data, this dataframe computes the average team score using formula,
    avg_score = (points_scored - points_allowed)/num_games_played
    
    Args:
        data: A python dictionary containing data in the following format.
            Winner has either TeamX or TeamY based on who wins the game.
            {'Team1':
                {'Team2': [Team1Score, Team2Score, Winner],
                 'Team3': [Team1Score, Team3Score, Winner],
                  ...
                }
             'Team2':
                {'Team1': [Team1Score, Team2Score, Winner],
                 'Team3': [Team1Score, Team3Score, Winner],
                  ...
                }
             ...
            }

    Returns:
        A dictionary containing average team scores.
    """
    p = dict()
    for team1, value in data.iteritems():
        t1 = team1
        p[t1] = defaultdict(list)
        for team2, stats in value.iteritems():
            t2 = team2
            t1_score = stats[0]
            t2_score = stats[1]
            winner = stats[2]
            p[t1][t2].append([t1_score, t2_score])
    for team1, value in data.iteritems():
        t1 = team1
        for team2, stats in value.iteritems():
            t2 = team2
            t1_score = stats[0]
            t2_score = stats[1]
            winner = stats[2]
            if p.has_key(t2):
                p[t2][t1].append([t2_score, t1_score])
    sumScores = {}
    for t1, val in p.iteritems():
        t1_score = 0
        t2_score = 0
        numGames = 0
        for t2, v in val.iteritems():
            numGames = numGames + len(v)
            for s in v:
                t1_score = t1_score + s[0]
                t2_score = t2_score + s[1]
        sumScores[t1] = [t1_score, t2_score, numGames]
    pNew = dict()
    for team, scores in sumScores.iteritems():
        pNew[team] = (scores[0] - scores[1])/scores[2]
    return pNew
```

```python
def get_yearly_winners(data):
    """Gets winner for a given year

    Given yearly data, this dataframe computes point difference using the
    formula,
    
    M.r = p
    where,
        M is a matrix of size #Teams X #Teams
        r is a vector of size #Teams.
        p is a vector of size #Teams.
    
    - M has the number of games played by each team on the diagonal, and the
      negation of the number of times each team played head-to-head on the
      off-diagonal spots.
    
    - r vector is a list of the average end-of-regular season point
      differentials for each team.
      ((points scored - points allowed) / total number of games played by the team)

    - p is our prediction vector. 

    Args:
        data: A python dictionary containing data in the following format.
            Winner has either TeamX or TeamY based on who wins the game.
            {'Team1':
                {'Team2': [Team1Score, Team2Score, Winner],
                 'Team3': [Team1Score, Team3Score, Winner],
                  ...
                }
             'Team2':
                {'Team1': [Team1Score, Team2Score, Winner],
                 'Team3': [Team1Score, Team3Score, Winner],
                  ...
                }
             ...
            }

    Returns:
        A tuple containing pandas dataframes representing point difference
        and average team score.
        
        For example: (pDiff, r)
            pDiff: Pandas dataframe containing point differences
            r: Pandas dataframe containing average team scores.
    """
    # Create a matrix M and vector r in pandas Dataframe.
    fin = dict()
    for team1, value in data.iteritems():
        t1 = team1
        fin[t1] = dict()
        if len(value.keys()) == 0:
            for otherTeams, value in data.iteritems():
                if t1 != otherTeams:
                    fin[t1][otherTeams] = np.NaN
        else:
            for team2, stats in value.iteritems():
                t2 = team2
                t1_score = stats[0]
                t2_score = stats[1]
                winner = stats[2]
                count = 0
                if fin.has_key(t2) and fin[t2].has_key(t1) and not np.isnan(fin[t2][t1]):
                    temp = fin[t2][t1] - 1
                    fin[t1][t2] = temp
                    fin[t2][t1] = temp
                else:
                    fin[t1][t2] = count - 1
    fin = pd.DataFrame(fin).to_dict()
    homeTeams = list(set(fin.keys()))
    awayTeams = []
    for key in homeTeams:
        awayTeams = awayTeams + fin[key].keys()
    awayTeams = list(set(awayTeams))
    allTeams = list(set(homeTeams + awayTeams))
    
    # Updating inverse games
    for t in allTeams:
        t1 = t
        for t in allTeams:
            t2 = t
            try:
                val1 = fin[t1][t2]
                val2 = fin[t2][t1]
            except:
                if fin.has_key(t1) and not fin[t1].has_key(t2):
                    fin[t1][t2] = 0
                else:
                    fin[t2][t1] = 0
                val1 = fin[t1][t2]
                val2 = fin[t2][t1]
            if not pd.isnull(val1):
                if not pd.isnull(val2):
                    if val1 == val2:
                        continue
                    else:
                        print "Aha check", t1, val1, t2, val2
                else:
                    fin[t2][t1] = val1
            if not pd.isnull(val2):
                if not pd.isnull(val1):
                    if val1 == val2:
                        continue
                    else:
                        print "Aha check", t2, val2, t1, val1
                else:
                    fin[t1][t2] = val2
    # Making NaNs 0
    for t in allTeams:
        t1 = t
        for t in allTeams:
            t2 = t
            val1 = fin[t1][t2]
            val2 = fin[t2][t1]
            if pd.isnull(val1):
                if pd.isnull(val2):
                    fin[t1][t2] = 0
                    fin[t2][t1] = 0     
    # Add diaginal values.
    for k, v in fin.iteritems():
        row_sum = abs(sum(v.values()))
        for k2, v2 in v.iteritems():
            if k == k2:
                fin[k][k2] = row_sum
    M = pd.DataFrame(fin)
    p = get_avg_team_score(data)
    r = pd.DataFrame.from_dict(p, orient='index').sort()
    P = M.dot(r)
    P.columns = ['point_diff']
    P = P.sort(columns='point_diff', ascending=False)
    pDiff = P.ix[:,['point_diff']]
    r.columns = ['avg_score']
    r = r.sort(columns='avg_score', ascending=False)
    return (pDiff, r)
```

```python
# Get winner for all the years
def get_winners(data):
    """Gets winner for all the years by calling get_yearly_winners(data[year])
    for every year's data.

    Args:
        data: A python dictionary containing data in the following format.
            Winner has either TeamX or TeamY based on who wins the game.
            {year1:
                {'Team1':
                    {'Team2': [Team1Score, Team2Score, Winner],
                     'Team3': [Team1Score, Team3Score, Winner],
                      ...
                    }
                 'Team2':
                    {'Team1': [Team1Score, Team2Score, Winner],
                     'Team3': [Team1Score, Team3Score, Winner],
                      ...
                    }
                 ...
                }
             year2:
                {'Team1':
                    {'Team2': [Team1Score, Team2Score, Winner],
                     'Team3': [Team1Score, Team3Score, Winner],
                      ...
                    }
                 'Team2':
                    {'Team1': [Team1Score, Team2Score, Winner],
                     'Team3': [Team1Score, Team3Score, Winner],
                      ...
                    }
                 ...
                }
             ...
            }

    Returns:
        A dictionary containing point differences and average scores for all years.
        The key contains the year and the value contains list of pandas dataframes
        that represent point differences and average team scores.
    """
    years = data.keys()
    result = dict()
    for year in years:
        pDiff, avgScore = get_yearly_winners(data[year])
        result[year] = [pDiff, avgScore]
    return result
```

```python
def get_all_year_top_teams(ans):
    """ Gets top 25 teams of each year
    
    Args:
        ans dictionary with all year rankings
    
    returns:
        topTeams dictionary with top 25 teams of each year
    """
    
    years = set(gamesDF.year)
    topTeams = dict()
    
    for year in years:
        d = dict()
        t = get_top_25(ans,year)
        for key, val in t.iterrows():
            team = key.lstrip()
            d[team] = val
        topTeams[year] = d
        
    
    return topTeams
```

```python
def check_accuracy(allYearTopTeams, actualResult):
    """ Checks the accuracy of the top 25 teams predicted by our model
    
    Args:
        allYear_top_teams - top 25 teams for each year
        actualResult - The Actual 25 teams as shown College Football Playoff Ranking 
        year - Year
        
    Returns:
        accuracy of our prediction(How many teams from our prediction are there in top 25)
    """
    
    accuracy = dict()
    years = set(gamesDF.year)
    for year in years:
        t = actualResult[actualResult['Year']==year]
        numCorrect = 0
        for index, val in t.Team.iteritems():
            if val in allYearTopTeams[year]:
                numCorrect = numCorrect + 1
        
        pctCorrect = numCorrect*100/float(25)
        accuracy[year] = [25, numCorrect, pctCorrect]
    
    return accuracy
```

## ERROR ANALYSIS OF BASELINE MODEL

The following modules plot the error distribution and on it they'll overlay useful statistics like

+ the estimated gaussian curve
+ mean
+ median
+ standard deviation
+ central 68% of data

**Note - ** Error is defined as error = actual rank of the team - rank predicted by our model

```python
def get_yearly_team_ranking(ans):
    """ Gets yearly ranking of all the teams for all the years on the basis of point diference.
    
    Args:
        ans dict which contains year wise all the team point score data
        
    Returns:
        yearlyRanking dict which contains year wise ranking of all the teams
        The yearly ranking dict looks like:
        {year:
            {team_name:[rank],
            ...
            team_name:[rank]}
        }
        year:
            {team_name:[rank],
            ...
            team_name:[rank]}
        }
            
    """
    yearlyRanking = dict()
    for k, v in ans.iteritems():
        year = k
        pDiff = v[0]
        pDiff['Rank'] = pDiff.rank(axis=0, numeric_only=None, method='max', na_option='keep', ascending=False, pct=False)
        yearlyRanking[year] = pDiff['Rank']
     
    return yearlyRanking    
```

```python
yearlyRanking = get_yearly_team_ranking(ans)
```

```python
def get_error(yearlyRanking, actualResult):
    """ Gets error for all the top 25 teams in actual list of top 25 teams
        error = actual rank of the team - rank predicted by our model
        
    Args:
        yearlyRanking dict and actualResult dict
        
    Returns:
        An err dict which contains a yearwise list value of errors
        The err dict looks like following:
        {year:[err1,err2...,err25]
            ...
         year:[err1,err2...,err25]}
    """
    err = dict()
    years = set(gamesDF.year)
    for year in years:
        t = actualResult[actualResult['Year']==year]
        errVal = list()
        for index, val in t.Team.iteritems():
            #val1 = ' ' + val
            try:
                e = t[t['Team']==val].Rank - yearlyRanking[year].loc[val]
            except:
                e = 1
            errVal.append(int(e))
        
        err[year] = errVal
    
    
    return err
```

```python
error = get_error(yearlyRanking, actualResult)
```

```python
def central_tendency_plot(point, ci, y, color, label):
    plt.plot(ci, [y, y], "-", color=color, linewidth=4, label=label)
    plt.plot(point, y, "o", color=color, markersize=10)
```

```python
def plot_error_dist(column, year):
    #d = column.tolist()
    d = column
    d = [0 if math.isnan(x) else x for x in d]
    m = mean(d)
    med = median(d)
    s = std(d)

    x = linspace(min(d), max(d), 1000)
    gauss = lambda x, m, s: (1 / (s * sqrt(2 * pi)) * exp(-0.5 * ((x - m) / s) ** 2))
    y = gauss(x, m, s)

    n, b, patches = plt.hist(d, 15, normed=True)
    bin_max = np.max(n)
    m_y = bin_max/5
    
    central_tendency_plot(m, [m - s, m + s], m_y, colors[1], "std dev")

    ci = np.percentile(d, [16, 84])
    central_tendency_plot(med, ci,  m_y - bin_max/10, colors[2], "68% CI")
    
    plt.legend()
    plt.plot(x, y, color=colors[4])
    plt.xlabel('Error')
    plt.ylabel('Normalized Frequency')
    title = str(year) + ' Error Distribution'
    plt.title(title)
    plt.show()
```

```python
def get_schedule(gamesDF, allGames=False):
    """Gets playoff game ordering.

    Args:
        gamesDF: A pandas dataframe containing games data with following columns.
            'game_id',
            'year',
            'week',
            'away_team',
            'home_team',
            'away_score',
            'home_score',
            'winner'

    Returns:
        A dictionary containing year as key and pandas dataframe containing playoff game
        ordering and scores as value.
    """
    years = set(gamesDF.year)
    gameSchedule = dict()
    for year in years:
        if allGames:
            yearlyPlayoff = gamesDF[gamesDF.year == year]
        else:
            yearlyPlayoff = gamesDF[gamesDF.year == year].tail(11)

        yearlyPlayoff['game_order'] = yearlyPlayoff.date.rank(ascending=True).order()
        cols = ['game_order', 'home_team', 'away_team', 'home_score', 'away_score', 'winner']
        yearlyPlayoff = yearlyPlayoff[cols]
        gameSchedule[year] = yearlyPlayoff
    return gameSchedule
```

```python
def yearly_game_simulator(gameSchedule, teamRankings, score='point_diff'):
    """Simulates games played on a particular season and returns a dataframe which contains
    information about our accuracy.

    Args:
        gameSchedule: A pandas dataframe containing NFL playoff gameplay ordering.
        teamRankings: A pandas dataframe containing team rankings from Baseline model.

    Returns:
        A pandas dataframe which contains model prediction and its correctness when compared
        to actual data.
    """
    stats = gameSchedule.copy(deep=True)
    for i, game in gameSchedule.iterrows():
        game_order = game[0]
        home_team = game[1]
        away_team = game[2]
        home_score = game[3]
        away_score = game[4]
        winner = game[5]
        if home_team == winner:
            winner_score = home_score
            loser_score = away_score
        else:
            loser_score = home_score
            winner_score = away_score
        try:
            home_team_rank = int(teamRankings.ix[home_team][score])
            away_team_rank = int(teamRankings.ix[away_team][score])
            if home_team_rank >= away_team_rank:
                predicted_winner = home_team
                predicted_score_diff = home_team_rank - away_team_rank
            else:
                predicted_winner = away_team
                predicted_score_diff = away_team_rank - home_team_rank
        except:
            home_team_rank = np.NAN
            away_team_rank = np.NAN
            predicted_winner = home_team
            predicted_score_diff = np.NAN
        if predicted_winner == winner:
            correct_prediction = True
        else:
            correct_prediction = False
        score_diff = winner_score - loser_score
        stats.ix[i,'predicted_winner'] = predicted_winner
        stats.ix[i,'correct_prediction'] = correct_prediction
        stats.ix[i,'home_team_rank'] = home_team_rank
        stats.ix[i,'away_team_rank'] = away_team_rank
        stats.ix[i,'predicted_score_diff'] = predicted_score_diff
        stats.ix[i,'score_diff'] = score_diff
        stats.ix[i,'error'] = predicted_score_diff - score_diff
    return stats
```

```python
def game_simulator(gameSchedule, teamRankings, score='point_diff'):
    """Simulates games played on all seasons and returns a dataframe which contains
    information about our accuracy.

    Args:
        gameSchedule: A dictionary containing year as key and a pandas dataframe
            containing NFL playoff gameplay ordering as value.
        teamRankings: A dictionary containing year as key and a pandas dataframe
            containing team rankings from Baseline model as value.

    Returns:
        A dictionary containing year as key and a pandas dataframe which contains
        model prediction and its correctness when compared to actual data as value.
    """
    allStats = dict()
    for year, schedule in gameSchedule.iteritems():
        if score == 'avg_score':
            rankings = teamRankings[year][1]
        elif score == 'point_diff':
            rankings = teamRankings[year][0]
        yearlyStats = yearly_game_simulator(schedule, rankings, score)
        allStats[year] = yearlyStats
    return allStats
```

```python
def check_accuracy(allStats):
    accuracy = dict()
    for year, df in allStats.iteritems():
        numGames = len(df.correct_prediction)
        numCorrect = sum(df.correct_prediction)
        pctCorrect = numCorrect*100/float(numGames)
        accuracy[year] = [numGames, numCorrect, pctCorrect]
    return accuracy
```
