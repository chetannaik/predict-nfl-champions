```python
def get_winner(home_score, away_score, home_team, away_team):
    """Finds out the winner of a game.
    
    Based of the score it decides who the winner of a game is.
  
    Args:
        home_score: An interger value which represents home team score.
        away_score: An interger value which represents home team score.
        home_team: A string representing the home team.
        away_team: A string representing the away team.
        
    Returns:
        A string which represents the winner of a given game.
    """
    if home_score > away_score:
        return home_team
    else:
        return away_team
```

```python
df = {}
for year in years:
    df[year] = newDF[newDF.year == year]
```

```python
def initialize(data):
    X = data.ix[:, intFeatures]
    y = data.ix[:, 'score_diff']

    model = sm.OLS(y, X, missing='drop').fit()
    return model
```

```python
def get_games_till_now(game):
    curYear = int(game.year)
    curGameId = int(game.game_id)
    histYears = [year for year in years if year <= curYear]
    
    e = []
    for year in histYears:
        if year == curYear:
            tmp = df[year]
            e.append(tmp[tmp.game_id < curGameId])
        else:
            e.append(df[year])
    allDF = pd.concat(e)
    return allDF
```

```python
def get_game_features(g, gamesTillNow):
    hTeam = g.home_team
    aTeam = g.away_team
    gameYear = int(g.year)

    games = []
    try:
        gamesNow = gamesTillNow[gameYear]
        for row, game in gamesNow.iterrows():
            rhTeam = game.home_team
            raTeam = game.away_team
            if ((rhTeam == hTeam) or (rhTeam == aTeam) or (raTeam == hTeam) or (raTeam == aTeam)):
                games.append(pd.DataFrame(game).T)
    except:        
        for row, game in gamesTillNow.iterrows():
            rhTeam = game.home_team
            raTeam = game.away_team
            if ((rhTeam == hTeam) or (rhTeam == aTeam) or (raTeam == hTeam) or (raTeam == aTeam)):
                games.append(pd.DataFrame(game).T)
        games = games[-10:]
    

    if len(games) == 0:
        games = []
        for row, game in gamesTillNow.iterrows():
            rhTeam = game.home_team
            raTeam = game.away_team
            if ((rhTeam == hTeam) or (rhTeam == aTeam) or (raTeam == hTeam) or (raTeam == aTeam)):
                games.append(pd.DataFrame(game).T)
        games = games[-10:]
    numGames = len(games)
    games = pd.concat(games)
    average = pd.DataFrame(games[intFeatures].mean().round()).T
    return (numGames, average[intFeatures])
```

```python
def get_updated_data(game):
    curYear = game.year
    curGameId = game.game_id    
    histYears = [year for year in years if year <= curYear][-2:]
    
    e = []
    for year in histYears:
        if year == curYear:
            tmp = df[year]
            e.append(tmp[tmp.game_id <= curGameId])
        else:
            e.append(df[year])
    allDF = pd.concat(e)
    return allDF
```

```python
def train_updated_model(data):
    X = data.ix[:, intFeatures]
    y = data.ix[:, 'score_diff']

    model = sm.OLS(y, X, missing='drop').fit()
    return model
```

```python
def run_model(model, gameFeatures):
    gameResult = int(model.predict(gameFeatures[intFeatures]).round())
    return gameResult
```

```python
initData = newDF[newDF.year == 2000]
model = initialize(initData)

mtemp = []
games = newDF[newDF.year != 2000]

for ix, game in games.iterrows():
        gamesTillNow = get_games_till_now(game)
        numGames, gameFeatures = get_game_features(game, gamesTillNow)
        gameResult = run_model(model, gameFeatures)
        data = get_updated_data(game)
        model = train_updated_model(data)
        mtemp.append(gameResult)
```

```python
def get_correct(score_diff, predicted_score):
    if ((predicted_score > 0 and score_diff > 0) or (predicted_score < 0 and score_diff < 0)):
        return 1
    else:
        return 0
```

```python
def central_tendency_plot(point, ci, y, color, label):
    plt.plot(ci, [y, y], "-", color=color, linewidth=4, label=label)
    plt.plot(point, y, "o", color=color, markersize=10)
```

```python
def plot_error_dist(column, year):
    d = column.tolist()
    d = [0 if math.isnan(x) else x for x in d]
    m = mean(d)
    med = median(d)
    s = std(d)

    x = linspace(min(d), max(d), 1000)
    gauss = lambda x, m, s: (1 / (s * sqrt(2 * pi)) * exp(-0.5 * ((x - m) / s) ** 2))
    y = gauss(x, m, s)
    
    fig, ax = plt.subplots(1,1)
    bins = len(set(d))
    n, b, patches = ax.hist(d, bins=bins, normed=True)
#     ax.set_xticks(np.arange(min(d), max(d)))
    bin_max = np.max(n)
    m_y = bin_max/5
    
    central_tendency_plot(m, [m - s, m + s], m_y, colors[1], "std dev")

    ci = np.percentile(d, [16, 84])
    central_tendency_plot(med, ci,  m_y - bin_max/10, colors[2], "68% CI")
    
    plt.legend()
    ax.plot(x, y, color=colors[4])
    plt.xlabel('Error')
    plt.ylabel('Normalized Frequency')
    title = 'Prediction Error Distribution'
    plt.title(title)
    plt.show()
    print "Mean: ", m
    print "Standard Deviation: ", s
```
