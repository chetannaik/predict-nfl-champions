```python
trainDF = pd.read_csv('trainLinCOL.csv', index_col=0)
testDF = pd.read_csv('testLinCOL.csv', index_col=0)
```

```python
trainDF['score_diff'] = trainDF.home_score - trainDF.away_score
testDF['score_diff'] = testDF.home_score - testDF.away_score
```

```python
trainDF = trainDF.sort(column='game_id').reset_index(drop=True)
testDF = testDF.sort(column='game_id').reset_index(drop=True)
```

```python
trainDF['1st Downs_diff'] = trainDF['1st Downs_home'] - trainDF['1st Downs_away']
trainDF['3rd Down Attempts_diff'] = trainDF['3rd Down Attempts_home'] - trainDF['3rd Down Attempts_away']
trainDF['3rd Down Conversions_diff'] = trainDF['3rd Down Conversions_home'] - trainDF['3rd Down Conversions_away']
trainDF['4th Down conversions_diff'] = trainDF['4th Down conversions_home'] - trainDF['4th Down conversions_away']
trainDF['Fumbles Lost_diff'] = trainDF['Fumbles Lost_home'] - trainDF['Fumbles Lost_away']
trainDF['Passing Completions_diff'] = trainDF['Passing Completions_home'] - trainDF['Passing Completions_away']
trainDF['Passing Yards_diff'] = trainDF['Passing Yards_home'] - trainDF['Passing Yards_away']
trainDF['Penalties_diff'] = trainDF['Penalties_home'] - trainDF['Penalties_away']
trainDF['Penalty Yards_diff'] = trainDF['Penalty Yards_home'] - trainDF['Penalty Yards_away']
trainDF['Rushing Attempts_diff'] = trainDF['Rushing Attempts_home'] - trainDF['Rushing Attempts_away']
trainDF['Rushing Yards_diff'] = trainDF['Rushing Yards_home'] - trainDF['Rushing Yards_away']
trainDF['Time of Possession_diff'] = trainDF['Time of Possession_home'] - trainDF['Time of Possession_away']
trainDF['4th Down Attempts_diff'] = trainDF['4th Down Attempts_home'] - trainDF['4th Down Attempts_away']
trainDF['Interceptions Thrown_diff'] = trainDF['Interceptions Thrown_home'] - trainDF['Interceptions Thrown_away']
trainDF['Passing Attempts_diff'] = trainDF['Passing Attempts_home'] - trainDF['Passing Attempts_away']
```

```python
testDF['1st Downs_diff'] = testDF['1st Downs_home'] - testDF['1st Downs_away']
testDF['3rd Down Attempts_diff'] = testDF['3rd Down Attempts_home'] - testDF['3rd Down Attempts_away']
testDF['3rd Down Conversions_diff'] = testDF['3rd Down Conversions_home'] - testDF['3rd Down Conversions_away']
testDF['4th Down conversions_diff'] = testDF['4th Down conversions_home'] - testDF['4th Down conversions_away']
testDF['Fumbles Lost_diff'] = testDF['Fumbles Lost_home'] - testDF['Fumbles Lost_away']
testDF['Passing Completions_diff'] = testDF['Passing Completions_home'] - testDF['Passing Completions_away']
testDF['Passing Yards_diff'] = testDF['Passing Yards_home'] - testDF['Passing Yards_away']
testDF['Penalties_diff'] = testDF['Penalties_home'] - testDF['Penalties_away']
testDF['Penalty Yards_diff'] = testDF['Penalty Yards_home'] - testDF['Penalty Yards_away']
testDF['Rushing Attempts_diff'] = testDF['Rushing Attempts_home'] - testDF['Rushing Attempts_away']
testDF['Rushing Yards_diff'] = testDF['Rushing Yards_home'] - testDF['Rushing Yards_away']
testDF['Time of Possession_diff'] = testDF['Time of Possession_home'] - testDF['Time of Possession_away']
testDF['4th Down Attempts_diff'] = testDF['4th Down Attempts_home'] - testDF['4th Down Attempts_away']
testDF['Interceptions Thrown_diff'] = testDF['Interceptions Thrown_home'] - testDF['Interceptions Thrown_away']
testDF['Passing Attempts_diff'] = testDF['Passing Attempts_home'] - testDF['Passing Attempts_away']
```

```python
cols = ['game_id','date','year','week','home_team','away_team','home_score','away_score','winner','score_diff','1st Downs_diff','3rd Down Attempts_diff','3rd Down Conversions_diff','4th Down conversions_diff','Fumbles Lost_diff','Passing Completions_diff','Passing Yards_diff','Penalties_diff','Penalty Yards_diff','Rushing Attempts_diff','Rushing Yards_diff','Time of Possession_diff','4th Down Attempts_diff','Interceptions Thrown_diff','Passing Attempts_diff']
```
