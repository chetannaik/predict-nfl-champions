```python
def get_all_year_data():
    """ Reads the CSV files and convert them into pandas dataFrame in a specified format.
   
    Args: 
        CSV files of previous year (2002 to 2013)
    
    Returns:
        gamesDF dataframe consisting of all the data from 2002 to 2013"""
    
    # Reading the csv files from the specified folder
    path = r'/media/ankit/Stuffs/Courses/CSE591 - Data Science/college_data/college_data'
    allFiles = glob.glob(path + "/*.csv")
    
    fileList = []
    
    for f in allFiles:
        df = pd.read_csv(f)
        week = 1
        df.rename(columns={'Date':'date',
                           'Home Team':'home_team',
                           ' Vis Team': 'away_team',
                           ' Score.1': 'home_score',
                           ' Score': 'away_score'}, inplace=True)

        # Creating Years Column
        def get_year(col):
            return col.split('/')[-1]

        df['year'] = df['date'].apply(get_year)
        t =  min(df.year)
        df['year'] = int(t)
        df['week'] = week
        week = week+1
        # changing data type pf home_score and away_score
        def change_type(col):
            try:
                col = float(col)
            except:
                col = 0
            return col
        
        df['home_score'] = df['home_score'].apply(lambda x: change_type(x))
        df['away_score'] = df['away_score'].apply(lambda x: change_type(x))
        
        # Creating winner column 
        def get_winner(home_score, away_score, home_team, away_team):
            if home_score > away_score:
                return home_team
            else:
                return away_team

        df['winner'] = df.apply(lambda x: get_winner(x['home_score'], x['away_score'], x['home_team'], x['away_team']), axis=1)
            
        fileList.append(df)
    gamesDF = pd.concat(fileList,ignore_index=True)
    
    
    return gamesDF
```

```python
# Calling the function to create prevgamesDF 
prevGamesDF = get_all_year_data()
```

```python
def get_recent_season_data():
    """ Gets the gamesDF dataframe which contains the previous year play data and adds the 2014 data.
    
    Args:
        CSV files with 2014 games information
    
    Returns: 
        gamesDF with same coloumns but having the 2014 data.
    """
    path = r'/media/ankit/Stuffs/Courses/CSE591 - Data Science/College Football Data/cfblines2014'
    allFiles = glob.glob(path + "/*.csv")
    
    fileList = []
    for f in allFiles:
        df = pd.read_csv(f)
        
        # Changing the header names
        df.rename(columns={'Date':'date',
                           'Home Team':'home_team',
                           ' Vis Team': 'away_team',
                           ' Score': 'home_score',
                           ' Vscore': 'away_score'}, inplace=True)
            
        # Creating Years Column
        df['year'] = 2014
        
        # changing data type pf home_score and away_score
        def change_type(col):
            try:
                col = float(col)
            except:
                col = 0
            return col
        
        df['home_score'] = df['home_score'].apply(lambda x: change_type(x))
        df['away_score'] = df['away_score'].apply(lambda x: change_type(x))
        
        # Creating winner column 
        def get_winner(home_score, away_score, home_team, away_team):
            if home_score > away_score:
                return home_team
            else:
                return away_team

        df['winner'] = df.apply(lambda x: get_winner(x['home_score'], x['away_score'], x['home_team'], x['away_team']), axis=1)
            
        fileList.append(df)
    gamesDF = pd.concat(fileList, ignore_index=True)
   
    
    return gamesDF
```

```python
intFeatures = [' Rushing Yards',' Rushing Attempts',' Passing Yards','Passing Attempts',' Passing Completions',' Penalties',' Penalty Yards',' Fumbles Lost','Interceptions Thrown',' 1st Downs',' 3rd Down Attempts',' 3rd Down Conversions','4th Down Attempts',' 4th Down conversions',' Time of Possession',' Rushing Yards.1',' Rushing Attempts.1',' Passing Yards.1','Passing Attempts.1',' Passing Completions.1',' Penalties.1',' Penalty Yards.1',' Fumbles Lost.1','Interceptions Thrown.1',' 1st Downs.1',' 3rd Down Attempts.1',' 3rd Down Conversions.1','4th Down Attempts.1',' 4th Down conversions.1',' Time of Possession.1']
```

```python
years = unique(newDF.year.astype(int))
years
```

```python
df = {}
for year in years:
    df[year] = newDF[newDF.year == year]
```

```python
def initialize(data):
    X = data.ix[:, intFeatures]
    y = data.ix[:, 'score_diff']

    model = sm.OLS(y, X, missing='drop').fit()
    return model
```

```python
def get_games_till_now(game):
    curYear = int(game.year)
    curGameId = int(game.game_id)
    histYears = [year for year in years if year <= curYear]
    e = []
    for year in histYears:
        if year == curYear:
            tmp = df[year]
            e.append(tmp[tmp.game_id < curGameId])
        else:
            e.append(df[year])
    allDF = pd.concat(e)
    return allDF
```

```python
def get_game_features(g, gamesTillNow):
    hTeam = g.home_team
    aTeam = g.away_team
    games = []

    for row, game in gamesTillNow.iterrows():
        rhTeam = game.home_team
        raTeam = game.away_team
        if (rhTeam == hTeam and raTeam == aTeam) or (rhTeam == aTeam and raTeam == hTeam):
            games.append(pd.DataFrame(game).T)

    if len(games) == 0:
        games = []
        for row, game in gamesTillNow.iterrows():
            #print game
            rhTeam = game.home_team
            raTeam = game.away_team
            if ((rhTeam == hTeam) or (rhTeam == aTeam) or (raTeam == hTeam) or (raTeam == aTeam)):
                games.append(pd.DataFrame(game).T)

    numGames = len(games)
    #print numGames
    games = pd.concat(games)
    average = pd.DataFrame(games[intFeatures].mean().round()).T
    return (numGames, average[intFeatures])
```

```python
def get_updated_data(game):
    curYear = game.year
    curGameId = game.game_id    
    histYears = [year for year in years if year <= curYear]
    
    e = []
    for year in histYears:
        if year == curYear:
            tmp = df[year]
            e.append(tmp[tmp.game_id <= curGameId])
        else:
            e.append(df[year])
    allDF = pd.concat(e)
    return allDF
```

```python
def train_updated_model(data):
    X = data.ix[:, intFeatures]
    y = data.ix[:, 'score_diff']

    model = sm.OLS(y, X, missing='drop').fit()
    return model
```

```python
def run_model(model, gameFeatures):
    gameResult = int(model.predict(gameFeatures[intFeatures]).round())
    return gameResult
```

```python
initData = newDF[newDF.year == 2005]
model = initialize(initData)

mtemp = []
games = newDF[newDF.year != 2005]
#years3 = [2006, 2007]
#l = list()
#for year in years3:
    #l.append(newDF[newDF.year == year])
    #games = pd.concat(l)

for ix, game in games.iterrows():
        gamesTillNow = get_games_till_now(game)
        numGames, gameFeatures = get_game_features(game, gamesTillNow)
        gameResult = run_model(model, gameFeatures)
        data = get_updated_data(game)
        model = train_updated_model(data)
        mtemp.append(gameResult)
```

```python
def get_correct(score_diff, predicted_score):
    if ((predicted_score > 0 and score_diff > 0) or (predicted_score < 0 and score_diff < 0)):
        return 1
    else:
        return 0
```

```python
process['correct_prediction'] = process.apply(lambda x: get_correct(x['score_diff'], x['predicted_score_diff']), axis=1)
```

```python
def check_accuracy(allStats):
    accuracy = dict()
    for year, wdict in allStats.iteritems():
        weekAccuracy = dict()
        for week, df in wdict.iteritems():
            numGames = len(df.correct_prediction)
            numCorrect = sum(df.correct_prediction)
            pctCorrect = numCorrect*100/float(numGames)
            weekAccuracy[week] = [numGames, numCorrect, pctCorrect]
        accuracy[year] = weekAccuracy
    return accuracy
```
